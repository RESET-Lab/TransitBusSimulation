# Gradient Querying

Gradient information is calculated for each segment of each bus route using the formula:

$$ \frac{Δelevation}{{distance}} $$

The unit for both measurements is meters so the result is unitless. Elevation data is aquired from the Google Maps Elevation API  which can be found here: https://developers.google.com/maps/documentation/elevation

Given the short length of certain segments and noise in the elevation data, it was observed that certain gradients were outliers (as depicted in the "Gradient Data for Raw Data" figure). Hence, the top and bottom percentile of the data was imputed to the 99 and 1 percentile gradient respectively.

The final result can be seen in the last figure.

```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt
import tqdm

import numpy as np

import requests
from bs4 import BeautifulSoup
import json

from shapely.geometry import LineString

import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
```

```python
PROJECTED_PATH = '../Data/Bus Routes/Projected'
RAW_OUTPUT_PATH = '../Data/Gradient Data/Raw'
ADJUSTED_OUTPUT_PATH = '../Data/Gradient Data/Adjusted'
```

```python
file_names = os.listdir(PROJECTED_PATH)
bus_route_files = []
bus_route_frames = []

for file in file_names:
    if ".shp" in file:
        bus_route_files.append(file)
        
fig, ax = plt.subplots(1, 1,figsize=(50,25))
ax.axis('off')
for file in tqdm.tqdm(bus_route_files,"Plotted"):
    cur = gpd.read_file(os.path.join(PROJECTED_PATH,file)) 
    bus_route_frames.append(cur)  
    cur.plot(ax = ax)
```

```python
API_KEY = "AIzaSyC9ERZN7gr_VosQdPbDNHJ8EAaU16h0P64"
URL = "https://maps.googleapis.com/maps/api/elevation/json?locations="

NUM_QUERIES = 0

def getElevations(coords):
    global NUM_QUERIES
    ans = []
    max_query = 300
    for i in range(0,len(coords),max_query):
        NUM_QUERIES += 1
        tq.set_postfix({'API Calls': NUM_QUERIES})
        query = coords[i:min(len(coords),i+max_query)]
        query_URL =  URL + str(coords[0][1]) + "," + str(coords[0][0])
        for point in query[1::]:
            query_URL += "|" + str(point[1]) + "," + str(point[0])
        query_URL += "&key=" + str(API_KEY)
        r = requests.get(query_URL, verify=False)
        data = json.loads(r.text)
        for point in data['results']:
            ans.append(point['elevation'])
    return ans
```

```python
nodes = []
data_points = []
edges_frames = []

tq = tqdm.tqdm(range(len(bus_route_frames)),"Querying")

for i in tq:
    frame = bus_route_frames[i]
    conv_frame = bus_route_frames[i].to_crs("EPSG:4326")
    name = bus_route_files[i].replace(".shp","")
    cur_edges = gpd.GeoDataFrame()
    cur_geom = []
    cur_grades = []
   
    for k in range(len(frame['geometry'])):
        geom = frame['geometry'][k]
        geom_conv = conv_frame['geometry'][k]

        elevations = getElevations(geom_conv.coords)

        geometries = []
        grades = []
        for j in range(len(geom.coords)-1):
            line = LineString([(geom.coords[j][0],geom.coords[j][1]),(geom.coords[j+1][0],geom.coords[j+1][1])])
            if(line.length == 0):
                continue
            geometries.append(line)
            grades.append((elevations[j+1] - elevations[j])/line.length)
        
        cur_geom.extend(geometries) 
        cur_grades.extend(grades)

        
    cur_edges['geometry'] = cur_geom
    cur_edges['grade'] = cur_grades
    
    edges_frames.append(cur_edges)
    
    cur_edges.to_file(os.path.join(RAW_OUTPUT_PATH,bus_route_files[i]))
    
```

```python
outputs = []

for path in os.listdir(RAW_OUTPUT_PATH):
    if ".shp" in path:
        outputs.append(path)
        
full_df = gpd.GeoDataFrame()
for path in outputs:
    path = os.path.join(RAW_OUTPUT_PATH,path)
    df = gpd.read_file(path)
    full_df = full_df.append(df)

fig, ax = plt.subplots(1,2,figsize = (15, 5))

ax[0].hist(np.log(np.abs(full_df['grade']) + 0.01))
ax[0].set_xlabel('log(abs(grade) + 0.01)')
ax[0].set_ylabel('Occurences')

ax[1].plot(sorted(full_df['grade']))
ax[1].set_xlabel('Sorted Position')
ax[1].set_ylabel('Grade')

fig.suptitle('Gradient Analysis for Raw Data')
pass
```

```python
n = len(full_df['grade'])
bottom = sorted(full_df['grade'])[n//100]
top = sorted(full_df['grade'])[n - n//100]


for file in tqdm.tqdm(outputs,"Processing"):
    path = os.path.join(RAW_OUTPUT_PATH,file)
    out_path = os.path.join(ADJUSTED_OUTPUT_PATH,file)
    
    df = gpd.read_file(path)
    
    df['grade'] = np.maximum(df['grade'],bottom)
    df['grade'] = np.minimum(df['grade'],top)

    df.to_file(out_path) 
```

```python
full_df = gpd.GeoDataFrame()
for path in tqdm.tqdm(outputs,"Reading"):
    path = os.path.join(ADJUSTED_OUTPUT_PATH,path)
    df = gpd.read_file(path)
    full_df = full_df.append(df)
```

```python
fig, ax = plt.subplots(1,2,figsize = (15, 5))

ax[0].hist(full_df['grade'])
ax[0].set_xlabel('Grade')
ax[0].set_ylabel('Occurences')

ax[1].plot(sorted(full_df['grade']))
ax[1].set_xlabel('Sorted Position')
ax[1].set_ylabel('Grade')

fig.suptitle('Gradient Analysis')
```
